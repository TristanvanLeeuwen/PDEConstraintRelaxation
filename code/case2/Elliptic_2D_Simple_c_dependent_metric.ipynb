```python
msh = mesh.create_rectangle(comm=MPI.COMM_WORLD,
                            points=((0.0, 0.0), (1.0, 1.0)), n=(16, 16),
                            cell_type=mesh.CellType.triangle)
V = fem.functionspace(msh, ("Lagrange", 1))

facets = mesh.locate_entities_boundary(msh, dim=(msh.topology.dim - 1),
                                       marker=lambda x: np.logical_or.reduce((
                                           np.isclose(x[0], 0.0),
                                           np.isclose(x[0], 1.0),
                                           np.isclose(x[1], 0.0),
                                           np.isclose(x[1], 1.0))))
dofs = fem.locate_dofs_topological(V=V, entity_dim=1, entities=facets)
bc = fem.dirichletbc(value=ScalarType(0), dofs=dofs, V=V)
```

```python
# create boundary points, starting from (0,0), clockwise
n = 11;
x = np.linspace(0, 1, n)
c = np.zeros((4*n-4,2))
c[n:2*n-1,0] = x[1:]
c[2*n-1:3*n-3,0] = 1
c[3*n-3:,0] = np.flip(x[1:])
c[0:n,1] = x
c[n:2*n-2,1] = 1
c[2*n-2:3*n-3,1] = np.flip(x[1:])
c
plt.scatter(c[:,0], c[:,1]) 
```

```python
###  The source/receiver functions near the boundaries
### Note this is the Rizes Representation of the true sources!!!
f_all = []
for i in range(40):
    f = fem.Function(V)
    f.interpolate(lambda x: np.exp(-(pow(x[0] - c[i,0] ,2)+pow(x[1] - c[i,1],2))*20) )
    f_all.append(f)
```

```python
u_trial = ufl.TrialFunction(V)
u_test = ufl.TestFunction(V)

x = ufl.SpatialCoordinate(msh)

atrue = (ufl.sin(x[0]*3)**2 + ufl.sin(x[1]*3)**2) +(ufl.sin(10*x[0])**2 + ufl.sin(10*x[1])**2)

M = np.zeros((40,40))
A = np.zeros((40,40))

# we used the standard H^1 semi norn inner product

for j in range(40):
    for i in range(40):
        M[i,j] = fem.assemble_scalar(fem.form( inner( atrue * grad(f_all[j]), grad(f_all[i]))*dx ) )
        # A[i,j] = fem.assemble_scalar(fem.form( inner( atrue * grad(f_all[i]), grad(f_all[j])) * dx ))

# D = M*np.linalg.inv(A)*M
# because M = A
D = M
```

```python
nk = 51
obj_all_inf = np.zeros(nk,)
obj_all_zero = np.zeros(nk,)
obj_all_small = np.zeros(nk,)
obj_all_large = np.zeros(nk,)

c_all = np.linspace(0, 2, nk)


for k in range(nk):
    # c = (k-(nk-1)/2)*20
    c = c_all[k]
    a = ufl.sin(x[0]*3)**2 + ufl.sin(x[1]*3)**2 + (1+100*c)*ufl.sin(10*x[0])**2 + ufl.sin(10*x[1])**2
    
    A = D*0 # initialize matrix A
    for j in range(40):
        for i in range(40):
            A[i,j] = fem.assemble_scalar(fem.form(  inner(a * grad(f_all[i]), grad(f_all[j])) * dx ) )
    A_inv = np.linalg.inv(A)
    
    # B = A
    G = A
    E = D - A 
    obj_all_inf[k] =  0.5*np.trace(E.transpose()*E)
    obj_all_large[k] =  0.5*np.trace(E.transpose()* np.linalg.inv(np.eye(40) + G/1e-3)*E)
    obj_all_small[k] =  0.5*np.trace(E.transpose()* np.linalg.inv(np.eye(40) + G/1e-4)*E)
    obj_all_zero[k] =  0.5*np.trace(E.transpose()*A_inv*E)
```

```python
font = FontProperties()
font.set_family('serif')
# font.set_name('Times New Roman')
font.set_size('12')
# font.set_style('italic')

coeffs  = c_all

fig, ax =  plt.subplots(figsize=(4, 3))

ax.plot(coeffs, obj_all_inf/np.max(obj_all_inf),color='r',label =  r'$\rho =\infty$')
ax.plot(coeffs, obj_all_large/np.max(obj_all_large),color='g',label = r'$\rho$ large')
ax.plot(coeffs, obj_all_small/np.max(obj_all_small),color='k',label = r'$\rho$ small')
ax.plot(coeffs, obj_all_zero/np.max(obj_all_zero),color='b',label = r'$\rho =0$')
ax.legend()
ax.set_xlabel('coefficient', fontproperties=font)
ax.set_ylabel('normalized obj. function', fontproperties=font)
plt.savefig('Poisson2D-source-basis-C-H1semi-inner.pdf',bbox_inches='tight')  
```

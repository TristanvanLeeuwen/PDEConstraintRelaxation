```python
import matplotlib.pyplot as plt
from mpi4py import MPI
from petsc4py.PETSc import ScalarType 
import numpy as np
import scipy
from matplotlib.font_manager import FontProperties

import ufl
from dolfinx import fem, io, mesh, plot
from dolfinx.fem import assemble_scalar, form
from dolfinx.fem.petsc import LinearProblem
from ufl import ds, dx, grad, inner
```

```python
msh = mesh.create_rectangle(comm=MPI.COMM_WORLD,
                            points=((0.0, 0.0), (1.0, 1.0)), n=(32, 32),
                            cell_type=mesh.CellType.triangle)
V = fem.functionspace(msh, ("Lagrange", 1))
```

```python
facets = mesh.locate_entities_boundary(msh, dim=(msh.topology.dim - 1),
                                       marker=lambda x: np.logical_or.reduce((
                                           np.isclose(x[0], 0.0),
                                           np.isclose(x[0], 1.0),
                                           np.isclose(x[1], 0.0),
                                           np.isclose(x[1], 1.0))))
dofs = fem.locate_dofs_topological(V=V, entity_dim=1, entities=facets)
bc = fem.dirichletbc(value=ScalarType(0), dofs=dofs, V=V)
```

```python
# create boundary points, starting from (0,0), clockwise
n = 11;
x = np.linspace(0, 1, n)
c = np.zeros((4*n-4,2))+ x[0]
c[n:2*n-1,0] = x[1:]
c[2*n-1:3*n-3,0] = x[-1]
c[3*n-3:,0] = np.flip(x[1:],axis = 0)
c[0:n,1] = x
c[n:2*n-2,1] = x[-1]
c[2*n-2:3*n-3,1] = np.flip(x[1:],axis = 0)
# c
plt.scatter(c[:,0], c[:,1]) 

plt.savefig('schrodinger_basis_loc.png', bbox_inches = 'tight')

```

```python
c_k = np.random.rand(10,1)
nx = 100
x = np.linspace(0, 1, nx)
print(c_k)
def atrue_normal(x,c_k):
    out = 0
    for k in range(10):
        out = out + c_k[k]*(np.sin(x[0]*(k+1))**2 + np.sin(x[1]*(k+1))**2)
    return out
atrue_handle = lambda x: atrue_normal(x,c_k)
atrue = fem.Function(V)
atrue.interpolate(atrue_handle)  
a_true = np.zeros((nx,nx))
for i in range(nx):
    for j in range(nx):
        a_true[i,j] = atrue_normal((x[i],x[j]),c_k)

plt.figure(figsize=(10, 3))
plt.imshow(a_true, cmap='RdBu', vmin=0, vmax=8, extent=[0, 1, 0, 1], interpolation='none')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.colorbar()
plt.savefig('schrodinger_a_true.png', bbox_inches = 'tight')
```

```python
###  The source/receiver functions near the boundaries
#### Note these f is the source term belonging to the weaker space
f_all = []
s = 3
for i in range(40):
    f = fem.Function(V)
    f.interpolate(lambda x: np.exp(-(pow(x[0] - c[i,0] ,2)+ \
                                     pow(x[1] - c[i,1],2))*(s**2))*s)
    f_all.append(f)
```

```python
u_trial = ufl.TrialFunction(V)
u_test = ufl.TestFunction(V)
x = ufl.SpatialCoordinate(msh)

# atrue = ufl.sin(x[0])**2 + ufl.sin(x[1])**2 + ufl.sin(10*x[0])**2 + ufl.sin(10*x[1])**2  

atrue = fem.Function(V)
# atrue.interpolate(lambda x: 1 + ( (x[0]-0.5)**2 + (x[1] - 0.5)**2 < 0.01) + \
#                                 (pow(np.sin(x[0]*10),2) + pow(np.sin(x[1]*10),2)))
psi_all = []
for k in range(10):
    psi = fem.Function(V)
    psi.interpolate( lambda x: np.sin(x[0]*(k+1))**2 + np.sin(x[1]*(k+1))**2 )
    psi_all.append(psi)
```

```python
# obtain p_i's from f_i's
## Note that p_i is the Riesz representation of f_i based on a_true!!!
p_all = []
B = inner(grad(u_trial),grad(u_test))*dx 
for i in range(40):
    RHS = f_all[i] * u_test * dx
    problem = LinearProblem(B, RHS, bcs=[bc], petsc_options={"ksp_type": "preonly", "pc_type": "lu"})
    u = problem.solve()
    p_all.append(u)
```

```python
M = np.zeros((40,40))
S = M
K = np.zeros((40*40,10))
for j in range(40):
    for i in range(j+1):
        M[i,j] = assemble_scalar(form(inner(grad(p_all[i]),grad(p_all[j]))*dx ))
        S[i,j] = assemble_scalar(form(inner(p_all[i], p_all[j])*dx))
        for k in range(10):
            K[j*40+i,k] = assemble_scalar( \
                    form(inner( psi_all[k]*p_all[i], p_all[j])*dx))
            K[j*40+i,k] = K[i*40+j,k]
        S[j,i] = S[i,j]
        M[j,i] = M[i,j]
```

```python
l = 10
A = M*0 
for j in range(40):
    for i in range(j+1):
        A[i,j] = assemble_scalar(form( inner(grad(p_all[i]),grad(p_all[j]))*dx \
            + inner(atrue*p_all[i],p_all[j])*dx \
            - inner(l*p_all[i], p_all[j])*dx ))
        A[j,i] = A[i,j]
```

```python
# W = M @ np.linalg.inv(D) @ M 
W = A
RHS = - M.flatten(order = 'F') + l* S.flatten(order = 'F')  + W.flatten(order = 'F') 
```

```python
K1 = K[0:40,:]
KK = K1.T@K1
new_RHS = RHS[0:40]
c_inv = scipy.linalg.lstsq(K1,new_RHS)[0]
```

```python
c_inv = c_inv/np.max(c_inv)
c_inv
```
